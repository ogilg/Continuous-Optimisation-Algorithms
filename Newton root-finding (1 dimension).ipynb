```python
import matplotlib.pyplot as plt
import autograd.numpy as np
from autograd import grad
import seaborn
```

```python
#plots f between a and b (integers)
def plot(f,a,b):
    x = np.arange(a,b,(b-a)/50)
    y = f(x)
    fig, ax = plt.subplots()
    ax.plot(x,y)
    ax.grid(True, which='both')
    
    ax.set_xlabel("x axis")
    ax.set_ylabel("y axis")
    
    seaborn.despine(ax=ax, offset=0)

    return ax
```

```python
def f(x):
    return (x-3)*(x-7)
```

```python
tol = 0.00000001
```

```python
def Newton_step(f,d_f,x):
    return x - f(x)/d_f(x)
```

```python
from matplotlib.collections import LineCollection
def newton(f,a):
    d_f = grad(f)
    iters = 0
    while abs(f(a)) >= tol:
        old_a = a
        a = Newton_step(f,d_f,a)
        ax = plot(f,-20,30)
        l1 = [(old_a, f(old_a)), (a, 0)]
        lc = LineCollection([l1], color=["k"], lw=2)
        
        plt.gca().add_collection(lc)
        
        if iters > 10000:
            print("iteration limit reached")
            return
        else:
            iters +=1
            
    print("converged in %d iterations" %iters)
```
