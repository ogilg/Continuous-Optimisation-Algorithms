```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
```

```python
#plots f between a and b (integers)
def plot(f,a,b):
    x = np.arange(a,b,(b-a)/50)
    y = f(x)
    fig, ax = plt.subplots()
    ax.plot(x,y)
    ax.grid(True, which='both')
    
    ax.set_xlabel("x axis")
    ax.set_ylabel("y axis")
    
    seaborn.despine(ax=ax, offset=0)

    return ax
```

```python
def f(x):
    return (x-3)*(x-7)
```

```python
tol = 0.00000001
```

```python
def secant_step(f,prev_x,x):
    new =  x - f(x)*(x-prev_x)/(f(x)-f(prev_x))
    return (x,new)
```

```python
from matplotlib.collections import LineCollection
def secant(f,prev_a,a):
    iters = 0
    while abs(f(a)) >= tol:
        prev_a,a = secant_step(f,prev_a,a)
        ax = plot(f,-20,30)
        l1 = [(prev_a, f(prev_a)), (a, 0)]
        lc = LineCollection([l1], color=["k"], lw=2)
        
        plt.gca().add_collection(lc)
        
        if iters > 10000:
            print("iteration limit reached")
            return
        else:
            iters +=1
            
    print("converged in %d iterations" %iters)
```
