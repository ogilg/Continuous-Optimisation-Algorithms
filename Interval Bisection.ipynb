```python
import numpy as np
import seaborn
import matplotlib.pyplot as plt
```

```python
#plots f between a and b (integers)
def plot(f,a,b):
    x = np.arange(a,b,(b-a)/50)
    y = f(x)
    fig, ax = plt.subplots()
    ax.plot(x,y)
    ax.grid(True, which='both')
    
    ax.set_xlabel("x axis")
    ax.set_ylabel("y axis")
    
    seaborn.despine(ax=ax, offset=0)

    return ax
```

```python
tol = 0.000000001
```

```python
# finds root of function using interval bisection from starting bracket (a0,b0)
def intervalBisection(f,a,b):
    assert a < b
    assert f(a)*f(b) < 0
    iterations = 0
    if f(a) < 0:
        case = 'increasing'
    else:
        case = 'decreasing'
        
    while abs(a-b) > tol:
        assert f(a)*f(b) < 0
        mid = (a+b)/2; y_mid = f(mid)
        if abs(y_mid) <= tol:
            print(mid)
            return mid
        elif (f(mid) < 0 and case == 'increasing') or (f(mid) > 0 and case == 'decreasing'):
            a = mid
        else:
            b = mid
            
        if iterations > 10000:
            print("iteration limit reached")
            return (a+b)/2
        else:
            iterations += 1
            
        #if iterations % 1 == 0:
        ax = plot(f,a-abs(a+b),b+abs(a+b))
        
        ax.axvline(a,color='r')
        ax.axvline(b,color='r')
        plt.show()
        plt.close()
        
    return (a+b)/2

    
```

```python
def search(f,a,b):

    while(f(a)*f(b) >= 0):
        a *= 2; b *= 2
        if b > 10000000:
            return False
    return (a,b)

def findBracket(f):
    a = -1.; b = 1.
    while search(f,a,b) == False:
        a = a/10; b = b/10
        
    return search(f,a,b)
```

```python
def findRoot(f):
    (a,b) = findBracket(f)
    print("starting bracket is (%s,%s)" % (a,b))
    return intervalBisection(f,a,b)
```
